```python
import numpy as np
import matplotlib.pyplot as plt

# Read the helix sequence
with open("file.txt", "r") as file:
    genome_sequence = file.read()

genome_list = list(genome_sequence)
genome_length = len(genome_list)

# Calculate helix structure
t = np.linspace(0, 4 * np.pi, genome_length)
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length)
coordinates = np.column_stack((x, y, z))

color_map = {
    'A': 'red',
    'T': 'blue',
    'C': 'green',
    'G': 'yellow'
}
colors = [color_map[base] for base in genome_list]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(coordinates[:, 0], coordinates[:, 1], coordinates[:, 2], c=colors, marker='o',s=100)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Helix Structure')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import requests
from io import BytesIO

url = "https://www.google.com/images/branding/googlelogo/2x/googlelogo_color_272x92dp.png"
response = requests.get(url)
image = Image.open(BytesIO(response.content))

img_array = np.array(image)

plt.imshow(img_array)
plt.title("Original Image")
plt.axis('off')
plt.show()

rotated_img = np.rot90(img_array)
flipped_img = np.fliplr(img_array)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(rotated_img)
plt.title("Rotated Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(flipped_img)
plt.title("Flipped Image")
plt.axis('off')

plt.show()

# grayscale filter
gray_img = np.dot(img_array[..., :3], [0.299, 0.587, 0.114])

plt.imshow(gray_img, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = np.array(iris.data)
Y = np.array(iris.target)
feature_names = iris.feature_names

# Calculations
for i in range(X.shape[1]):
    print(f"Feature: {feature_names[i]}")
    print(f"  Mean: {np.mean(X[:, i])}")
    print(f"  Median: {np.median(X[:, i])}")
    print(f"  Standard Deviation: {np.std(X[:, i])}")
    print(f"  Minimum: {np.min(X[:, i])}")
    print(f"  Maximum: {np.max(X[:, i])}")
    print("-" * 20)

sepal_data = X[:, :2]
print("Sepal Data (Sepal Length, Sepal Width):\n", sepal_data)

# Visualizations
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='viridis')
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.title("Sepal Length vs Sepal Width")
plt.colorbar(label='Species')
plt.show()

plt.figure(figsize=(8, 6))
plt.hist(X[:, 0], bins=20, color='skyblue', edgecolor='black')
plt.xlabel(feature_names[0])
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Length")
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(X[:, 2], X[:, 3], marker='o', linestyle='-', color='coral')
plt.xlabel(feature_names[2])
plt.ylabel(feature_names[3])
plt.title("Petal Length vs Petal Width")
plt.grid(True)
plt.show()
```
